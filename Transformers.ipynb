```python
from transformers import pipeline

# Initialize the question-answering pipeline with a specific model
question_answer = pipeline('question-answering', model='distilbert/distilbert-base-cased-distilled-squad', device=-1)

# The story context (you can replace this with any text)
context = """
In the heart of a foggy old town stood a tiny clock shop, its windows dusted with time and its bell rarely rung. It was owned by Mr. Elric, a quiet man with silver hair, kind eyes, and a workshop full of ticking wonders. Nobody really knew where he came from or how he knew so much about time.

Twelve-year-old Leo passed the shop every day on his way to school. He loved clocks — the way they ticked in perfect rhythm, as if keeping secrets. One rainy afternoon, as lightning split the sky, Leo noticed the shop light was flickering. The door was slightly open.

Curious and cautious, Leo stepped inside.

“Hello?” he called, but no one answered.

The clocks ticked wildly, not in sync but in chaos. Then, as he reached the back room, he found something strange — a tall grandfather clock with glowing blue hands and no numbers. It wasn’t just a clock; it was a door.

Without thinking, Leo touched it — and the world shifted.

He was no longer in the clock shop. He stood in a land of frozen time, where birds hovered mid-flight and raindrops floated in the air. In the middle of it all stood Mr. Elric, calmly adjusting a giant gear in the sky.

"You’re not supposed to be here yet," Elric said with a smile. "But perhaps you're the one I’ve been waiting for."

From that day, Leo became his apprentice — a guardian of time.
"""

print("Ask a question about the story (type 'end' to stop):")

while True:
    user_question = input("Your question: ")

    if user_question.strip().lower() == 'end':
        print("Exiting the question-answering loop. Goodbye!")
        break

    result = question_answer(question=user_question, context=context)
    print("Answer:", result['answer'])
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load pretrained DialoGPT model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/DialoGPT-medium")
model = AutoModelForCausalLM.from_pretrained("microsoft/DialoGPT-medium")

# Chat history
chat_history_ids = None

# User input
user_input = "Hello, how are you?"

# Encode the user input and add end-of-string token
new_input_ids = tokenizer.encode(user_input + tokenizer.eos_token, return_tensors='pt')

# Append to chat history (or start fresh)
bot_input_ids = torch.cat([chat_history_ids, new_input_ids], dim=-1) if chat_history_ids is not None else new_input_ids

# Generate a response
chat_history_ids = model.generate(bot_input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)

# Decode and print
response = tokenizer.decode(chat_history_ids[:, bot_input_ids.shape[-1]:][0], skip_special_tokens=True)
print("Bot:", response)
```
